```python
from skimage.segmentation import watershed
from skimage.feature import local_binary_pattern
import numpy as np
import matplotlib.pyplot as plt
import cv2
from copy import deepcopy
from skimage.morphology import disk
from skimage.segmentation import watershed
from skimage import data
from skimage.filters import rank
from skimage.util import img_as_ubyte
from scipy import ndimage as ndi
```

```python
img = plt.imread('/content/bubbles.JPG', format = 'gray')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
blur = cv2.medianBlur(gray, ksize = 13)
plt.imshow(blur)
```

```python
wshed = watershed(255 - blur, watershed_line = True)
final_image = deepcopy(img)
final_image[wshed==0] = (255,0,0)
fig, ax = plt.subplots(figsize=(90, 10))
ax.imshow(final_image, interpolation='nearest')

plt.imshow(final_image)
```

```python
fig, ax = plt.subplots(figsize=(90, 10))
ax.imshow(wshed, interpolation='nearest')

plt.imshow(wshed)
```
